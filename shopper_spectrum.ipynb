# **Project Summary :**


**Objective:**

The principal objective of the Shopper Spectrum initiative is to segment customers based on their purchasing behavior and develop a recommendation engine to suggest relevant products. By leveraging historical transaction data, the project aims to uncover hidden patterns in customer behavior using unsupervised learning (RFM-based clustering) and item-based collaborative filtering techniques. This dual approach empowers e-commerce platforms to deliver personalized experiences, improve marketing strategies, and enhance customer retention.

**Dataset Overview**

The dataset consists of real-world e-commerce transaction data, containing both numerical and categorical features:

- InvoiceNo: Unique transaction identifier.

- StockCode and Description: Product identification and name.

- Quantity and UnitPrice: Number of items purchased and per-unit cost.

- InvoiceDate: Timestamp of the purchase.

- CustomerID: Unique customer identifier.

- Country: Customer's country of origin.

The dataset spans transactions from 2022 to 2023 and exhibits the typical challenges of missing values, duplicate invoices, and irregular transaction records.

**Data Preprocessing and Feature Engineering**

Comprehensive preprocessing steps were performed to ensure the dataset was analysis-ready:

- Removal of records with missing CustomerID.

- Exclusion of cancelled invoices (identified by InvoiceNo starting with "C").

- Filtering out transactions with negative or zero Quantity and UnitPrice.

- Conversion of InvoiceDate to datetime format for temporal analysis.

Subsequently, RFM (Recency, Frequency, Monetary) features were engineered for each customer:

- Recency: Days since the customer's last purchase.

- Frequency: Number of transactions by the customer.

- Monetary: Total amount spent.

These features were standardized using StandardScaler to enable effective clustering.

**Clustering and Customer Segmentation**

The standardized RFM data was used to segment customers via KMeans Clustering. The optimal number of clusters was determined using both the Elbow Method and Silhouette Score.

Each cluster was interpreted and labeled based on the average RFM values:

- High-Value: Recent, frequent, and high-spending customers.

- Regular: Moderate frequency and spending.

- Occasional: Rare purchasers with lower monetary value.

- At-Risk: Long inactive and low-value customers.

Cluster labels serve as a foundation for targeted marketing, loyalty programs, and churn prevention strategies.

**Recommendation System Development**

An Item-Based Collaborative Filtering model was implemented using cosine similarity. A customer-product interaction matrix was constructed, and similarity scores between products were calculated.

Given a product input, the model returns the top 5 most similar products based on historical co-purchases. This approach enhances user engagement by offering personalized product recommendations in real-time.

**Streamlit Application**

A web interface was developed using Streamlit to provide an interactive platform:

- Product Recommendation Module: Takes a product name as input and displays 5 similar products.

- Customer Segmentation Module: Accepts RFM values and predicts the customer segment label.

The UI is designed to be clean, intuitive, and business-facing, enabling real-time demonstration of the underlying models.

**Evaluation and Insights**

Model performance was validated using inertia and silhouette scores for clustering quality. Customer clusters were visualized via 2D and 3D scatter plots, while product similarity was showcased using heatmaps.

These insights revealed actionable patterns:

- Concentration of high spenders in a specific segment.

- Popular products frequently bought together.

- Purchase trends varying by country and time of year.

**Business Impact**

The project delivers tangible business benefits:

- Enables customer lifecycle targeting (retention, reactivation).

- Drives product upselling and cross-selling through recommendations.

- Assists in inventory optimization based on clustered demand.

- Informs dynamic pricing and loyalty initiatives.

By integrating customer segmentation with product recommendations, this project offers a holistic toolkit for data-driven decision-making in e-commerce.

Future enhancements may include:

**Chart 7 : Recency Distribution**

```python
# Convert `InvoiceDate` to datetime and `CustomerID` to integer
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['CustomerID'] = df['CustomerID'].astype(int)

# Calculate `TotalPrice`
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

# Define the latest date in the dataset
latest_date = df['InvoiceDate'].max()

# Calculate RFM metrics
rfm_df = df.groupby('CustomerID').agg(
    Recency=('InvoiceDate', lambda date: (latest_date - date.max()).days),
    Frequency=('InvoiceNo', lambda num: len(num.unique())),
    Monetary=('TotalPrice', 'sum')
).reset_index()

# Standardize the RFM values
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_df[['Recency', 'Frequency', 'Monetary']])
rfm_scaled_df = pd.DataFrame(rfm_scaled, columns=['Recency', 'Frequency', 'Monetary'])

# Apply KMeans clustering with 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
rfm_df['Cluster'] = kmeans.fit_predict(rfm_scaled_df)

# Calculate the mean RFM values for each cluster
cluster_profiles = rfm_df.groupby('Cluster')[['Recency', 'Frequency', 'Monetary']].mean()
cluster_profiles = cluster_profiles.reset_index()

# Melt the DataFrame for plotting
cluster_profiles_melted = cluster_profiles.melt('Cluster', var_name='Metric', value_name='Value')

# Create the visualization
plt.figure(figsize=(10, 6))
sns.barplot(x='Cluster', y='Value', hue='Metric', data=cluster_profiles_melted)
plt.title('Customer Cluster Profiles based on RFM Metrics')
plt.xlabel('Customer Cluster')
plt.ylabel('Mean Value (Non-standardized)')
plt.savefig('customer_cluster_profiles.png')
plt.show()
```

1. **Why this chart?**

To analyze the geographical distribution of revenue and identify high-performing countries, especially when one country (like the UK) dominates. The logarithmic scale is useful for making differences between smaller contributors more visible, which would otherwise be dwarfed by the primary contributor.


2. **Insight:**

The United Kingdom contributes an overwhelmingly large share of the total revenue.

Other top contributors, such as the Netherlands, EIRE, Germany, and France, generate significantly lower revenue volumes compared to the UK.

The revenue gap between the UK and other countries remains stark even with a logarithmic scale, highlighting the UK's disproportionate contribution.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.decomposition import PCA
```

**Step3 : Elbow Method to Find Optimal Clusters**

```python
wcss = []  # within-cluster sum of squares

for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmeans.fit(rfm_scaled)
    wcss.append(kmeans.inertia_)

# Plot WCSS vs number of clusters
plt.figure(figsize=(8, 5))
plt.plot(range(2, 11), wcss, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.tight_layout()
plt.savefig('visuals/elbow_method.png')
plt.show()
```

1. **Why this chart?**

To determine the optimal number of customer segments (k) for clustering.

2. **Insight:**

A clear "elbow" in the curve (usually at k = 3 or k = 4) shows the best balance between compact clusters and model simplicity.

3. **Business Impact:**

Helps create clear, non-overlapping customer groups for targeted marketing.


4. **Negative Insight:**

A lack of an elbow point may signal that the customer base is not easily segmentable based on RFM alone.

**Step4 : Applying KMeans**

We'll convert the unsupervised KMeans clusters from Phase 3 into labels for a classification problem.

**Step1: Import Required Libraries**

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

import joblib
```

```python
# -----------------------------------
# Product Recommendation Section
# -----------------------------------
if menu == "Product Recommendation":
    st.header("Product Recommendation Engine")
    st.markdown("Enter a product name to get recommendations for similar items.")

    # Input for product name
    product_name_input = st.text_input("Enter Product Name (e.g., WHITE HANGING HEART T-LIGHT HOLDER)")

    if st.button("Get Recommendations"):
        try:
            # Load the pre-processed data and similarity matrix
            # Ensure these paths are correct relative to where app.py is run
            df_path = os.path.join(data_dir, "cleaned_retail_data.csv") # Assuming this is your main data file
            similarity_matrix_path = os.path.join(models_dir, "product_similarity.pkl") # Assuming you saved this

            if not os.path.exists(df_path):
                st.error(f"Error: Data file not found at {df_path}. Please ensure it's present.")
            if not os.path.exists(similarity_matrix_path):
                st.error(f"Error: Similarity matrix file not found at {similarity_matrix_path}. Please ensure it's saved correctly.")

            # Load data and similarity matrix
            df_rec = pd.read_csv(df_path)
            df_rec['Description'] = df_rec['Description'].str.strip() # Clean descriptions
            item_similarity_matrix = joblib.load(similarity_matrix_path)

            # Function to get recommendations
            def get_recommendations(product_description, similarity_matrix, df_data, num_recommendations=5):
                # Ensure the product description exists in the similarity matrix index
                if product_description not in similarity_matrix.index:
                    return f"Product '{product_description}' not found in the recommendation system. Please check the spelling or try another product."

                # Get similarity scores for the given product
                sim_scores = similarity_matrix[product_description]

                # Sort products by similarity score in descending order
                sim_scores = sim_scores.sort_values(ascending=False)

                # Get the top N similar products (excluding the product itself)
                top_similar_products = sim_scores.index[1:num_recommendations+1] # Skip the first one (itself)

                return top_similar_products.tolist()

            # Get and display recommendations
            recommendations = get_recommendations(product_name_input.upper(), item_similarity_matrix, df_rec) # Convert to upper to match common product descriptions

            if isinstance(recommendations, str):
                st.warning(recommendations) # Display warning if product not found
            elif recommendations:
                st.subheader("Recommended Products:")
                for i, product in enumerate(recommendations):
                    st.write(f"{i+1}. {product}")
            else:
                st.info("No recommendations found for this product.")

        except Exception as e:
            st.error("An error occurred during recommendation. Please ensure your data and similarity matrix files are correct.")
            st.code(str(e))
```
